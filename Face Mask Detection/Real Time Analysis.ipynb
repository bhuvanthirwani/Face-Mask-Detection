```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
from PIL import Image
from io import BytesIO
import requests
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img,probability,flag):
    
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    probstr=str(probability)
    for (x,y,w,h) in face_rects:
        if(flag==1):
            cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,255,0), 10)
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(face_img,text=probstr,org=(x,y), fontFace=font,fontScale=1,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
        else:
            cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,0,255), 10)
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(face_img,text=probstr,org=(x,y), fontFace=font,fontScale=1,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
        print(flag)
    return face_img
```

```python
model = load_model('model.h5')
# summarize model.
#model.summary()
```

```python
def Probability(img,classifier):
    img=np.expand_dims(img,axis=0)
    img=img/255
    return 100*classifier.predict(img)[0][0],classifier.predict_classes(img)[0][0]
```

```python
cap = cv2.VideoCapture(0) 

while True: 
    
    ret, frame = cap.read(0) 
    #trial_img=image.load_img(frame,target_size=(64,64))
    #prob,flag=Probability(trial_img,model)
    #frame = detect_face(frame,prob,flag)
 
    fr = face_cascade.detectMultiScale(frame)
    if(len(fr)!=0):
        x,y,w,h=tuple(fr[0])
        print(x,y,w,h)
        img=frame[y:y+h,x:x+w]
        img = cv2.resize(img,(64, 64))
        print(img.shape,type(img))
        img2= Image.fromarray(img)
        print(type(img2))
        b = BytesIO()
        img2.save(b,format="jpeg")
        img3 = Image.open(b)
        print("HELLO",type(img3))
        prob,flag=Probability(img3,model)
        frame = detect_face(frame,prob,flag)
        print(prob,flag)
    cv2.imshow('Video Face Detection', frame)
    c = cv2.waitKey(1) 
    if c == 27: 
        break 
        
cap.release() 
cv2.destroyAllWindows()
```
